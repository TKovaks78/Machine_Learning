##### Simple imputer

**Documentation**:https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html#sklearn.impute.SimpleImputer

- **mean**: replace missing values using the mean along each column. Only for numerical values.
- **median**: replace missing values using the median along each column. Only for numerical values.
- **most_frequent**: replace missing using the most frequent value along each column. Can be used with strings or numeric data. If there is more than one such value, only the smallest is returned.
- **constant**: replace missing values with a constant value defined by the fill_value parameter. Can be used with strings or numeric data.

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')

# Fit the Imputer to the dataset
imp.fit(X)

# Transform (i.e., Impute the null values) the training dataset according to the learned statistics (i.e., mean of the column)
imp.transform(X)
```

<u>Sklearn Ordinal Encoding + missing values</u>

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.model_selection import cross_validate

#List the categorical columns on which you want to apply tranformation
categorical_columns = [
    "weather",
    "season",
    "holiday",
    "workingday",
]
#Map values for transformation
categories = [
    ['clear', 'misty', 'rain', 'heavy_rain'],
    ["spring", "summer", "fall", "winter"],
    ["False", "True"],
    ["False", "True"],
]

#Define encoder
ordinal_encoder = OrdinalEncoder(categories=categories)

#ColumnTranformer function is necessary to do the mapping
ct = ColumnTransformer(transformers=[("categorical", ordinal_encoder, categorical_columns)], remainder="passthrough")

#If you wish to apply change directly to you data

#Get values of the transformed data
#ordinal_values = ct.fit_transform(df)
#Implement to dataframe
#df = pd.DataFrame(ordinal_values, columns = list(df.columns))

#If you want seperate columns with ordinal dat

#Get values of the transformed data
ordinal_values = ct.fit_transform(df[categorical_columns])

#Implement data in the dataset
ordinal_cols = ["weather_ordinal", "season_ordinal", "holiday_ordinal","workingday_ordinal"]
df[ordinal_cols] = ordinal_values
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_validate

regressor = DecisionTreeRegressor()
cv_results_tree_regressor = cross_validate(
    regressor, X, y, cv=cv, scoring="neg_mean_absolute_error", n_jobs=2
)

errors_tree_regressor = pd.Series(
    -cv_results_tree_regressor["test_score"], name="Decision tree regressor"
)
errors_tree_regressor.describe()
```

```python
from sklearn.dummy import DummyRegressor

dummy = DummyRegressor(strategy="mean")
result_dummy = cross_validate(
    dummy, X, y, cv=cv, scoring="neg_mean_absolute_error", n_jobs=2
)
errors_dummy_regressor = pd.Series(
    -result_dummy["test_score"], name="Dummy regressor"
)
errors_dummy_regressor.describe()
```

```python
all_errors = pd.concat(
    [errors_tree_regressor, errors_dummy_regressor],
    axis=1,
)
all_errors.head()
```

- **K-Fold Cross Validation**: 
    - <font color="green"> Simple method, good start for regression problem </font>
    - <font color="red"> Do not use if distribution of target variable is not consistent </font>
- **Stratified K-Fold**
    - <font color="green"> Use in case of imbalanced target variable</font>
- **Leave-One-Out Cross Validation**
    - <font color="green"> Good for small datasets</font>
    - <font color="red"> may require large computation time </font>
- **Group K-fold Cross Validation**
    - <font color="green"> ensures the same group is not represented in the train and test set</font>
    - <font color="red"> Do not use outside this specific case</font>
- **Stratified K-fold Cross Validation**
    - <font color="green"> ensures the same group is not represented in the train and test set AND preserve class distribution</font>
    - <font color="red"> Do not use outside this specific case</font>
